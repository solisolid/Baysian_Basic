# 3.1　学習と予測

　本章では２章で紹介した基本的な確率分布を使って，ベイズ学習の基本であるパラメータの事後分布および未観測値の予測分布の解析的な計算を解説します．ここで紹介する典型的な計算例は，後の章で登場するような，より複雑な確率モデルに対する推論を行うための基礎となります．

## 3.1.1　パラメータの事後分布

　訓練データを$\ \mathcal{D}\ $としたとき，ベイズ学習では次のような**同時分布**$\ p(\mathcal{D},\theta)\ $を考えることによってデータを表現するモデルを考えます．

$$ p(\mathcal{D},\theta) = p(\mathcal{D} \mid \theta)p(\theta)$$

ここで，$\ \theta \ $はモデルに含まれる未知のパラメータです．パラメータに関する事前の不確実性は，**事前分布**$\ p(\theta)\ $を設定することによってモデルに反映されます．また，特定のパラメータ$\ \theta \ $からどのようにしてデータ$\ \mathcal{D} \ $が発生するのかを記述している部分である$\ p(\mathcal{D} \mid \theta)\ $を**尤度関数（likelihood function）**と呼びます．

$$ p(\theta \mid \mathcal{D}) = \frac{p(\mathcal{D} \mid \theta) p(\theta)}{p(\mathcal{D})}$$

この条件付き分布$\ p(\theta \mid \mathcal{D})\ $を計算することが，ベイズ学習の枠組みにおける「学習」にあたる部分になります．

## 3.1.2　予測分布

　さらに，学習されたパラメータの分布を使って，未観測のデータ$\ x_{*}\ $に対して何らかの知見を得たい場合があります．これは，次のような**予測分布（predictive distribution）**$\ p(x_{*} \mid \mathcal{D})\ $を計算することによって実現できます（証明は割愛させていただきます）．

$$ p(x_{*} \mid \mathcal{D}) = \int p(x_{*} \mid \theta)p(\theta \mid \mathcal{D}) d\theta $$

以上をまとめると，モデルを使って$\ x_{*}\ $を予測するためには，はじめに式を使ってパラメータの事後分布$\ p(\theta \mid \mathcal{D})\ $を学習し，得られた事後分布から式を用いて積分計算を行うという手順が必要になります．

## 3.1.3　共役事前分布

　事後分布や予測分布を効率良く計算するための方法として，**共役事前分布（conjugate prior）**を使うアイデアがあります．共役事前分布とは，式における事前分布$\ p(\theta)\ $と事後分布$\ p(\theta \mid \mathcal{D})\ $が同じ種類の確率分布を持つように設定された事前分布を指します．それゆえ，どのような事前分布が共役になりうるのかというのは，尤度関数$\ p(\mathcal{D} \mid \theta)\ $の設計の仕方によります．
