# 3.3　１次元ガウス分布の学習と予測

## 3.3.1　平均が未知の場合

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline

#学習用データ作成のためのメソッド（Chapter2_Lesson2で実装済み）
def multitrial_norm(mu, lam, n):
    result = []
    for i in range(n):
        result.append(norm.rvs(mu, 1 / lam))
    return result

#事後分布の可視化メソッド
def visual_posterior_norm_mu(lam_known, data, m, lam_mu):
    lam_mu_ = len(data) * lam_known + lam_mu
    m_ = (lam_known * np.sum(data) + lam_mu * m) / lam_mu_
    
    x = np.arange(0, 10, 0.001)
    y = norm.pdf(x, m_, 1 / lam_mu_)
    plt.plot(x, y, label = "$m=$" + str(m) + ", " + "$\lambda_{\mu}=$" + str(lam_mu))
    plt.title("Posterior distribution")
    plt.xlabel("$\mu$")
    plt.legend()
    
#予測分布の可視化メソッド
def visual_predictive_norm(lam_known, data, m, lam_mu):
    lam_mu_ = len(data) * lam_known + lam_mu
    m_ = (lam_known * np.sum(data) + lam_mu * m) / lam_mu_
    lam_ast = lam_known * lam_mu_ / (lam_known + lam_mu_)
    mu_ast = m_
    x = np.arange(-20, 20, 0.001)
    y = norm.pdf(x, mu_ast, 1 / lam_ast)
    plt.plot(x, y, label = "$m=$" + str(m) + ", " + "$\lambda_{\mu}=$" + str(lam_mu))
    plt.title("Predictive distribution")
    plt.xlabel("x")
    plt.ylabel("probability")
    plt.legend()
```

```python
#実行
mu_truth = 5
lam_known = 0.2
N = 10
data = multitrial_norm(mu_truth, lam_known, N)
print(data)
print(np.mean(data))

visual_posterior_norm_mu(lam_known, data, 1.0, 1.0)
visual_posterior_norm_mu(lam_known, data, 2.0, 0.5)
visual_posterior_norm_mu(lam_known, data, 5.0, 0.1)
```

```python
visual_predictive_norm(lam_known, data, 1.0, 1.0)
visual_predictive_norm(lam_known, data, 2.0, 0.5)
visual_predictive_norm(lam_known, data, 5.0, 0.1)
```

## 3.3.2　精度が未知の場合

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.special import gamma
%matplotlib inline

#学習用データ作成のためのメソッド（Chapter2_Lesson2で実装済み）
def multitrial_norm(mu, lam, n):
    result = []
    for i in range(n):
        result.append(norm.rvs(mu, 1 / lam))
    return result

#事後分布の可視化メソッド
def visual_posterior_norm_lambda(mu_known, data, a, b):
    def gamma_distribution(la, a, b):
        return b**a / gamma(a) * la**(a-1) * np.e**(-b*la)
    variance_data = []
    for i in range(len(data)):
        variance_data.append(1 / 2 * (data[i] - mu_known) ** 2)
    a_ = len(data) / 2 + a
    b_ = np.sum(variance_data) + b
    x = np.arange(0, 10, 0.001)
    y = gamma_distribution(x, a_, b_)
    plt.plot(x, y, label = "a=" + str(a) + ", " + "b=" + str(b))
    plt.title("Posterior distribution")
    plt.xlabel("$\lambda$")
    plt.legend()
    
#予測分布の可視化メソッド
def visual_predictive_norm_lambda(mu_known, data, a, b):
    mu_s = mu_known
    lam_s = a / b
    nu_s = 2 * a
    x = np.arange(-3, 3, 0.001)
    y = gamma((nu_s+1)/2) / gamma(nu_s/2) * (lam_s/np.pi/nu_s) ** (1/2) * (1+lam_s/nu_s*(x-mu_s)**2) ** (-(nu_s+1)/2)
    plt.plot(x, y, label = "a=" + str(a) + ", " + "b=" + str(b))
    plt.title("Predictive distribution")
    plt.xlabel("x")
    plt.ylabel("probability")
    plt.legend()
```

```python
#実行
mu_known = 0
lam_truth = 2.0
N = 10
data = multitrial_norm(mu_known, lam_truth, N)
print(data)
print(np.mean(data))

visual_posterior_norm_lambda(mu_known, data, 1.0, 1.0)
visual_posterior_norm_lambda(mu_known, data, 2.0, 0.5)
visual_posterior_norm_lambda(mu_known, data, 3.0, 0.1)
```

```python
visual_predictive_norm_lambda(mu_known, data, 1.0, 1.0)
visual_predictive_norm_lambda(mu_known, data, 2.0, 0.5)
visual_predictive_norm_lambda(mu_known, data, 3.0, 1.5)
```
